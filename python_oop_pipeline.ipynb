####Sobre o Projeto
Repositório desenvolvido durante o curso "Pipeline de dados: combinando Python e orientação a objeto" da Alura.

####Objetivo
Construir um pipeline de dados reutilizável aplicando Programação Orientada a Objetos (POO).

####O Cenário
Simulação da fusão de duas empresas que precisam integrar suas bases de dados de vendas (em formatos diferentes) em um único arquivo consolidado e limpo.
